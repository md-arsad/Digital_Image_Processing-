#Code for Erosion

```python
def erosion(image, structuring_element, cen):
    
    # output_image = np.zeros_like(image)
    # Get the dimensions of the structuring element
    se_rows, se_cols = structuring_element.shape
    x,y=cen[0],cen[1]
    rs,cs=structuring_element.shape
    # Get the dimensions of the input image
    rows, cols = image.shape
    # Initialize the output image
    output_image = np.zeros((rows,cols))
    # Loop over each pixel in the input image
    for i in range(rows):
        for j in range(cols):
            # Check if the pixel in the input image matches the reference point
            if image[i, j] == 1:
                u1=x        # u1 row back from  current  row
                d1=rs-1-x   # d1 row forward from current row
                u2,d2=y,cs-1-y   #u2 col back and d2 col forward from current col
                s=(i-u1,j-u2)
                if(i-u1<0 or i+d1>rows-1 or j-u2<0 or j+d2>cols-1):  # check if structuring element go out of boundary
                    continue
                flag=True
                for m in range(i-u1,i+d1+1):           # checking whther structuring element find same structure in the image or not
                    for n in range(j-u2,j+d2+1):
                        if(structuring_element[m-s[0],n-s[1]]==1):
                            if(image[m,n]!=1):
                                flag=False
                                break 
                    if flag==False:
                      break
                if flag==True:
                  output_image[i,j]=1
    sx1=x                            #since in Dilation we have given the padding of size(structuring_element)-1   
    sx2=rs-x-1                        # there for we are removing the padding 
    sy1=y
    sy2=cs-y-1
    r,c=output_image.shape
    crop_img=np.zeros(((r-rs)+1,(c-cs)+1)) 
    crop_img= output_image[sx1:r-sx2,sy1:c-sy2]
    return crop_img
```

#Code for Dilation

```python
def Dilation(image, st_ele, cen):
   
    # Get the dimensions of the structuring element
    se_rows, se_cols = st_ele.shape
    x,y=cen[0],cen[1]
    rs,cs=st_ele.shape
    # Get the dimensions of the input image
    rows, cols = image.shape
    # sx1=x
    # sx2=rs-x-1
    # sy1=y
    # sy2=cs-y-1
    # print(s1,s2)
    # Initialize the output image
    output_image = np.zeros((rows+rs-1,cols+cs-1))
    
    
    # print("initial output") 
    # print(output_image)
    # Loop over each pixel in the input image
    
    for i in range(x,rows+x):
        for j in range(y,cols+y):
            # Check if the pixel in the input image matches the reference point
            if image[i-x, j-y] == 1:
                u1=x                       # u1 row back from  current  row
                d1=rs-1-x                  # d1 row forward from  current  row
                u2,d2=y,cs-1-y             # u2 col back and d2 col forward from current col
                s=(i-u1,j-u2)       
                # if(i-u1<0 or i+d1>rows-1 or j-u2<0 or j+d2>cols-1):
                #     continue
                
                for m in range(i-u1,i+d1+1):
                    for n in range(j-u2,j+d2+1):
                        # print(st_ele[m-s[0],n-s[1]],end=",")
                        if(st_ele[m-s[0],n-s[1]]==1):
                            output_image[m,n]=1                   
    return output_image
```

#Flow Chart of ersosion Detector
How does the erosion  detector works ?

The erosion detector takes in as image_to_check , structuring_element , and reference point as the arguements.

The image to check is  a binary image since we are building detectors for detecting morphological filtering in binary images

Our paper has used the most common structuring elements which are frequently used in image filtering .

So the structuring element which is passed must be any one of the given atructuring elements in this python notebook.

In the next step , we perform closing operation on the image to check . 

If the image to check was eroded earlier , performing the closing operation on the eroded image with the same structuring element with which it was eroded gives back the eroded image .

So we can compare the image to check to with the image we get after performing closing operation on it .

If both are equal, we return true

Our paper has used the most common structuring elements which are frequently used in image filtering .

In the next step , we perform opening morphological filter on the image_to_check.

If the image has been eroded , the opening operation with the same structuring element with which it was eroded will return the same eroded image .

So , now we comapare the image_to_check with the image we get after performing the opening operation . If both are equal ,we return true .
